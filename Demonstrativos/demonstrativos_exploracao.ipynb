```python
import pandas as pd
import os
import numpy as np
```

# PETR4

## Info inicial dos dados raw

```python
df = pd.read_csv(os.path.join("Tables","PETR4.csv"))
df.info()
```

```python
df.head()
```

```python
df.drop(0, inplace= True)
```

```python
df = df.T
df.columns = df.loc["Descrição"]
df.drop("Descrição", inplace= True)
print(df.info())
df.head()
```

## Apagando colunas nulas

```python
colunas_nulas = ["Dívida Líquida/Ebitda","Resultado não operacional - (R$)","CAPEX - (R$)","Dívida Líquida - (R$)", "ROE - (%)", "ROIC - (%)"                        ]
df[colunas_nulas]
```

```python
df.drop(columns= colunas_nulas,inplace= True)
```

## Apagando linhas redundantes

AV é o valor total dividido pela receita Líquida. Consideramos esse dado redundante

```python
colunas_av = [coluna for coluna in df.index if "AV" in coluna]
df.drop(index= colunas_av, inplace= True)
```

## Convertendo Valores

```python
def converter_valor(valor):
    if "M" in valor:
        return int(valor.replace(".","").replace(",00", "").replace(" M","000"))
    if "%" in valor: 
        return float(valor.replace('.','').replace(",",".").replace("%",""))/100
```

```python
df["Margem Bruta - (%)"].apply(converter_valor)
```

```python
for coluna in df.columns:
    df[coluna]= df[coluna].apply(converter_valor)
    
df
```

```python
df.info()
```

## separando colunas AH
(medem variação em relação ao último resultado)

```python
indices_ah = [indice for indice in df.index if "AH" in indice]
df_ah = df.loc[indices_ah]
df.drop(index= indices_ah, inplace= True)
```

## Salvando com DateTime

```python

def trimestre_para_datetime(trimestre_str):
    # Extrai o número do trimestre e o ano da string
    numero_trimestre = int(trimestre_str[0])
    ano = int(trimestre_str[2:])
    
    # Mapeia o número do trimestre para o primeiro mês do trimestre
    primeiro_mes_do_trimestre = (numero_trimestre - 1) * 3 + 1
    
    # Cria um objeto datetime usando o ano e o primeiro mês do trimestre
    data = pd.to_datetime(f'{ano}-{primeiro_mes_do_trimestre:02d}-01')
    
    return data

trimestre = '4T2023'
data = trimestre_para_datetime(trimestre)
print(data)
```

```python
df.index = pd.Series(df.index).apply(trimestre_para_datetime)
df.index
```

```python
# excluimos o ultimo pois nao podemos comparar o primeiro com o que veio antes, pela propria definicao de primeiro
df_ah.index = df.index[:-1].copy()
```

```python
df = df//1000
df = df.astype(int)
```

```python
df.to_csv("Tables\Limpas\PETR4.csv")
df_ah.to_csv("Tables\Limpas\PETR4_AH.csv")
```

# Alguma estatistica

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,5))
plt.title("Receita Líquida",fontdict= {"size":40})
for acao in ["PETR4","VALE3"]:
    df = pd.read_csv('Tables\Limpas\\'+acao+'.csv')
    df.rename(columns= {"Unnamed: 0": "Data"}, inplace= True)
    df.index = df["Data"]
    df.drop(columns= ["Data"], inplace= True)
    
    plt.bar(df["Receita Líquida - (R$)"].index,df["Receita Líquida - (R$)"].values /1000,label= acao)

plt.xlabel("data",fontdict= {"size":40})
plt.ylabel("Milhões R$",fontdict= {"size":40})
plt.legend()
plt.show()
```
